**Bisection Method:**

<span class="underline">Introduction:</span>

The bisection method is one of the techniques to find the root value of
the polynomial equation. This method follows the principle of
intermediate theorem which states “*An equation f(x)=0 is a real
continuous function, has at least one root between* $x_{l}$ *and*
$x_{u}$ *if f (*$x_{l}$*)f(*$x_{u}$*)\<0.* So, in this method we shall
take two points as an initial guess such that both the points are used
in function and product of those function is a negative number.

$x_{l} = first\ initial\ guess$.

$x_{u} = second\ initial\ guess$.

<span class="underline">Algorithm:</span>

1.  Choose two points as an initial guess. Consider these initial guess
    as $x_{l}$ and $x_{u}$ such that f($x_{l}$)f($x_{u}$)\<0 this means
    f(x) changes their sign between $x_{l}$ and $x_{u}$.

2.  Now find the midpoint ($x_{m}$) of the points as $x_{m}$ =
    $\frac{x_{l} + x_{u}}{2}$ and find the value of function at $x_{m}$.

3.  Then check the following

<!-- -->

1.  If f ($x_{l}$)f($x_{m}$)\<0 then the root is in between
    $x_{l}\text{\ and\ }x_{m}$. Thus
    $x_{l} = x_{l}\text{\ and}\text{\ x}_{m} = \ x_{u}$.

2.  If $f(x_{l})f\left( x_{m} \right) > 0$ then the root value is in
    between
    $x_{m}\text{\ and\ }x_{u}\text{\ thus\ }x_{l} = \ x_{m}\text{\ and\ }x_{u} = \ x_{u}$.

3.  If $f(x_{l})f(x_{m}) = 0$ then the root value is $x_{m}$.

<!-- -->

1.  Now find the absolute relative approximation error after every
    iteration as

$\left| E_{a} \right| = \left| \frac{x_{m}^{\text{new}} - x_{m}^{\text{old}}}{x_{m}^{\text{new}}} \right| \times 100$

where

$x_{m}^{\text{new}} = current\ estimate\ value\ of\ the\ root$.

$x_{m}^{\text{old}} = previous\ estimate\ value\ of\ the\ root$.

1.  Compare the absolute relative approximate error$|E_{a}|$ with the
    given tolerance $E_{s}$.

-   If $\left| E_{a} \right| > E_{s}$ then repeat the algorithm from
    2<sup>nd</sup> step.

-   If $\left| E_{a} \right| < E_{s}$ then stop the algorithm.

<span class="underline">Advantages:</span>

-   It is convergent.

-   Root bracket gets halved with each iteration.

<span class="underline">Disadvantages:</span>

-   This method is slow convergence.

-   If one of the initial guesses is close to the root, the convergence
    is slower.

-   If function f (x) does not touches the x-axis then it is unable to
    find the root e.g. $f\left( x \right) = x^{2}$

-   Function changes sign but root does not exist
    e.g.$f\left( x \right) = \frac{1}{x}$

<span class="underline">Example:</span>

$$f\left( x \right) = x^{3} - 7x^{2} + 14x - 6$$

Let $0 \leq x \leq 1$

So,

$$x_{l} = 0\ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ x_{u} = 1$$

$$f\left( x_{l} \right) = f\left( 0 \right) = - 6$$

$$f\left( x_{u} \right) = f\left( 1 \right) = 2$$

As, $f\left( x_{l} \right)f\left( x_{u} \right) < 0$ so according to
theorem there exist at least one root between 0 and 1.

<span class="underline">Iteration 1:</span>

$x_{m} = \frac{x_{l} + x_{u}}{2}$.

$x_{m} = \frac{0 + 1}{2}$.

= 0.5

$f\left( 0.5 \right) = - 0.625$.

$f\left( 0 \right)f\left( 0.5 \right) > 0$.

$x_{l} = 0.5,x_{u} = 1$.

Hence, the root lies between 0.5 and 1

<span class="underline">Iteration 2:</span>

$f\left( 0.75 \right) = 0.9844$.

As, $f\left( 0.5 \right)f\left( 0.75 \right) > 0$

So,$x_{l} = 0.5\ \ \ \ \ \ \ \ \ \ x_{u} = 0.75$

Hence, the root lies between 0.5 and 0.75.

<span class="underline">Absolute Relative approximation error:</span>

$$\left| E_{a} \right| = |\frac{x_{m}^{\text{new}} - x_{m}^{\text{oid}}}{x_{m}^{\text{new}}}| \times 100$$

$$\left| E_{a} \right| = \left| \frac{0.75 - 0.5}{0.75} \right| \times 100 = 33.33\%$$

The number of significant digits at least correct is given by the
largest value or m for which

$$E_{a} \leq 0.5 \times 10^{2 - m}$$

$$0.3333 \leq 0.5 \times 10^{2 - m}$$

$$0.6666 \leq 10^{2 - m}$$

$$\log\left( 0.6666 \right) \leq \log\left( 10^{2 - m} \right)$$

$$- 0.1761 - 2 \leq - m$$

$$m = 2$$

So, number of significant digits at least correct in the root at the end
is 2.

**False position method:**

<span class="underline">Introduction:</span>

False position method is also called regula falsi method is another
technique of solving equation in one variable. In this method we start
finding the root by taking two initial guesses.

**Derivation:**

$$\frac{f(x_{l})}{x_{r} - x_{l}} = \frac{f(x_{u})}{x_{r} - x_{u}}$$

Where,

$x_{l} =$1<sup>st</sup> initial guess.

$x_{u} =$2<sup>nd</sup> initial guess.

$x_{r} = root\ value$.

$${(x}_{r} - x_{u})f(x_{l}) = f(x_{u})(x_{r} - x_{l})$$

$$x_{r}f\left( x_{l} \right) - x_{u}f\left( x_{l} \right) = x_{r}f\left( x_{u} \right) - x_{l}f\left( x_{u} \right)$$

$$x_{r}(f\left( x_{l} \right) - f\left( x_{u} \right)) = x_{u}f\left( x_{l} \right) - x_{l}f(x_{u})$$

$$x_{r} = \frac{x_{u}f\left( x_{l} \right) - x_{l}f(x_{u})}{f\left( x_{l} \right) - f(x_{u})}$$

Is the required formula for calculating the root value in false position
method.

<span class="underline">Algorithm:</span>

1.  Choose two points as an initial guess. Consider these initial guess
    as $x_{l}$ and $x_{u}$.

$$x_{r} = \frac{x_{u}f\left( x_{l} \right) - x_{l}f(x_{u})}{f\left( x_{l} \right) - f(x_{u})}$$

1.  Then check the following

<!-- -->

1.  If f ($x_{l}$)f($x_{r}$)\<0 then the root is in between
    $x_{l}\text{\ and\ }x_{r}$. Thus
    $x_{l} = x_{l}\text{\ and}\text{\ x}_{r} = \ x_{u}$.

2.  If $f(x_{l})f\left( x_{r} \right) > 0$ then the root value is in
    between
    $x_{r}\text{\ and\ }x_{u}\text{\ thus\ }x_{l} = \ x_{r}\text{\ and\ }x_{u} = \ x_{u}$.

3.  If $f(x_{l})f(x_{r}) = 0$ then the root value is $x_{r}$.

<!-- -->

1.  Now find the absolute relative approximation error after every
    iteration as

$\left| E_{a} \right| = \left| \frac{x_{r}^{\text{new}} - x_{r}^{\text{old}}}{x_{r}^{\text{new}}} \right| \times 100$

where

$x_{r}^{\text{new}} = current\ estimate\ value\ of\ the\ root$.

$x_{r}^{\text{old}} = previous\ estimate\ value\ of\ the\ root$.

1.  Compare the absolute relative approximate error$|E_{a}|$ with the
    given tolerance $E_{s}$.

-   If $\left| E_{a} \right| > E_{s}$ then repeat the algorithm from
    2<sup>nd</sup> step.

-   If $\left| E_{a} \right| < E_{s}$ then stop the algorithm.

**Newton Raphson method:**

<span class="underline">Introduction:</span>

Newton Raphson method is one of the techniques of solving polynomial
equation. It is quadratically convergent when get approach to the root
value.

<span class="underline">Algorithm:</span>

1.  Find the first derivative of the function.

2.  Consider the initial guess $x_{i}$ and find the new estimate value
    as

$$x_{i + 1} = x_{i} - \frac{f(x_{i})}{f'(x_{i})}$$

1.  Now find the absolute relative approximation error after every
    iteration as

$\left| E_{a} \right| = \left| \frac{x_{i + 1} - x_{i}}{x_{i + 1}} \right| \times 100$

where

$x_{i + 1} = current\ estimate\ value\ of\ the\ root$.

$x_{i} = previous\ estimate\ value\ of\ the\ root$.

1.  Compare the absolute relative approximate error$|E_{a}|$ with the
    given tolerance $E_{s}$.

-   If $\left| E_{a} \right| > E_{s}$ then repeat the algorithm from
    2<sup>nd</sup> step.

-   If $\left| E_{a} \right| < E_{s}$ then stop the algorithm.

<span class="underline">Advantages:</span>

-   This method is converging fast to the root value.

-   At the start only one initial guess is required.

<span class="underline">Disadvantages:</span>

-   If the initial guess is selected near the inflection point of the
    function then the function starts diverges away from the root.

-   The initial value on which the value of the derivative of the
    function equal to zero are not allowed because division by zero is
    undefined.

-   If initial guess is chosen near the root value then sometimes it may
    jump to another root value.

<span class="underline">Example:</span>

Consider a function

$$f\left( x \right) = x^{3} - 7x^{2} + 14x - 6$$

$$f^{'}\left( x \right) = 3x^{2} - 14x + 14$$

<span class="underline">Graph:</span>



<span class="underline">Iteration 1:</span>

Let

$$x_{0} = 2$$

$$x_{1} = x_{0} - \frac{f(x_{0})}{f'(x_{0})}$$

Here,

$$f\left( x_{0} \right) = f\left( 2 \right) = 2$$

$$f^{'}\left( x_{0} \right) = f^{'}\left( 2 \right) = - 2$$

$$x_{i + 1} = 2 + \frac{2}{2} = 3$$

<span class="underline">Absolute approximation error:</span>

$\left| E_{a} \right| = \left| \frac{x_{1} - x_{0}}{x_{1}} \right| \times 100$

$$\left| E_{a} \right| = |\frac{3 - 2}{3}| \times 100$$

$$E_{a} = 33.33\%$$

**Secant method:**

<span class="underline">Introduction:</span>

The secant method is a technique in which a series of roots of secant
lines are used to find the root values of functions.

<span class="underline">Derivation:</span>

$$f'\left( x_{i} \right) = \frac{f\left( x_{i} \right) - f(x_{i - 1})}{x_{i} - x_{i - 1}}$$

$$x_{i + 1} = x_{i} - \frac{f(x_{i})}{\frac{f\left( x_{i} \right) - f(x_{i - 1})}{x_{i} - x_{i - 1}}}$$

$$x_{i + 1} = x_{i} - \frac{f(x_{i})(x_{i} - x_{i - 1})}{f\left( x_{i} \right) - f(x_{i - 1})}$$

<span class="underline">Algorithm:</span>

1.  Consider two initial guesses as $x_{i}$ and $x_{i - 1}$

2.  Now find the estimate value as

$$x_{i + 1} = x_{i} - \frac{f(x_{i})(x_{i} - x_{i - 1})}{f\left( x_{i} \right) - f(x_{i - 1})}$$

1.  Now find the absolute relative approximation error after every
    iteration as

$\left| E_{a} \right| = \left| \frac{x_{i + 1} - x_{i}}{x_{i + 1}} \right| \times 100$

where

$x_{i + 1} = current\ estimate\ value\ of\ the\ root$.

$x_{i} = previous\ estimate\ value\ of\ the\ root$.

1.  Compare the absolute relative approximate error$|E_{a}|$ with the
    given tolerance $E_{s}$.

-   If $\left| E_{a} \right| > E_{s}$ then repeat the algorithm from
    2<sup>nd</sup> step.

-   If $\left| E_{a} \right| < E_{s}$ then stop the algorithm.

<span class="underline">Advantages:</span>

-   This method is converging fast to the root value.

-   Without bracket the root it requires two initial guesses.

<span class="underline">Disadvantages:</span>

-   As division by zero is undefined so in this method if we such value
    in which subtraction of function on these values is equal to zero
    then it gives error.

-   If initial guess is chosen near the root value then sometimes it may
    jump to another root value.

**Fix point iteration method:**

<span class="underline">Introduction:</span>

A fixed point for a function is a number at which the function does
change its value when the function is applied. Here the number p is
considered as the fixed point of the function g.

<span class="underline">Convergence:</span>

The convergence of fixed point is based on existence and uniqueness
theorem. According to this theorem:

-   If g ∈ C \[a, b\] and g(x) ∈ \[a, b\] for all x ∈ \[a, b\], then g
    has at least one fixed point in \[a, b\].

<!-- -->

-   If, in addition, g’(x) exists on (a, b) and a positive constant k \<
    1 exists with \|g (x)\| ≤ k, for all x ∈ (a, b), then there is
    exactly one fixed point in \[a, b\].

> Then, for any number in \[a, b\], the sequence defined by

$$p_{n} = g(p_{n - 1})$$

> Converges to the unique fixed-point p in \[a, b\].
>
> <span class="underline">Algorithm:</span>

-   choose some function and make it in the form of
    $x = g\left( x \right)$ such that $g(x)$ is continuous.

-   Now take some initial guess as value of x.

-   Use x in the function $g\left( x \right)$.

-   Calculate the error after every iteration as

$$\left| E_{a} \right| = \left| \frac{x_{n} - x_{n - 1}}{x_{n}} \right|$$

> Where,

$x_{n} = current\ estimate\ value\ of\ the\ root$.

$x_{n - 1} = previous\ estimate\ value\ of\ the\ root$.

-   Compare the absolute relative approximate error$|E_{a}|$ with the
    given tolerance $E_{s}$.

<!-- -->

-   If $\left| E_{a} \right| > E_{s}$ then repeat the algorithm from
    3<sup>rd</sup> step.

-   If $\left| E_{a} \right| < E_{s}$ then stop the algorithm.

<span class="underline">Advantages:</span>

-   Fixed point iteration method is easier to solve.

-   It is a good root finding strategy.

-   Converges fast.

**Gauss Jacobi method:**

Jacobi method is one method to solve linear equation of the system. The
system written in form of matrix as Ax=b must have a unique solution and
have a non-zero diagonal entries. This method only possible when the is
strictly dominant which means the magnitude of the diagonal entries must
be greater than the sum of other entries of that row. If the matrix is
not dominant then make it dominant by interchanging the rows.

<span class="underline">Algorithm:</span>

-   Write system of equation in the form of Ax=b as

$$\begin{matrix}
a_{11}x_{1} & a_{12}x_{2}\cdots & a_{1n}x_{n} = b_{1} \\
a_{21}x_{1} & a_{22}x_{2}\cdots & a_{2n}x_{n} = b_{2} \\
\end{matrix}$$

$$\vdots \ \ \ \ \ \ \ \ \ \ \ \ \ \ \  \vdots \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \  \vdots \ \ \ \ \ \ \ \ \ \ \ \ $$

$$\begin{matrix}
a_{n1}x_{1} & a_{n2}x_{2}\cdots & a_{\text{nn}}x_{n} = b_{n} \\
\end{matrix}\text{\ \ \ \ }$$

-   We have to find the value of $x_{1},x_{2}\ldots x_{n}$ as

$$x_{1} = \frac{1}{a_{11}}(b_{1} - a_{12}x_{2}^{k - 1} - \ldots - a_{1n}x_{n}^{k - 1})$$

$$x_{2}^{k} = \frac{1}{a_{22}}\left( b_{2} - a_{21}x_{1}^{k - 1} - \ldots - a_{2n}x_{n}^{k - 1} \right)$$

Similarly,

$$x_{n}^{k} = \frac{1}{a_{\text{nn}}}\left( b_{n} - a_{1n}x_{1}^{k - 1} - a_{2n}x_{2}^{k - 1} - \ldots - a_{\text{nn}}x_{n}^{k - 1} \right)$$

-   Find the error of approximation as

$$E = \frac{\left| \left| x^{k} - x^{k - 1} \right| \right|}{\left| \left| x^{k} \right| \right|}$$

**Gauss-Seidel method:**

This method is similar to the Jacobi method but the only difference is
that it has use the updated values to find the other value for example,

$$x_{1}^{k} = \frac{1}{a_{11}}(b_{1} - a_{12}x_{2}^{k - 1} - \ldots - a_{1n}x_{n}^{k - 1})$$

$$x_{2}^{k} = \frac{1}{a_{22}}\left( b_{2} - a_{21}x_{1}^{k + 1} - \ldots - a_{2n}x_{n}^{k + 1} \right)$$

Similarly,

$$x_{n}^{k} = \frac{1}{a_{\text{nn}}}\left( b_{n} - a_{1n}x_{1}^{k - 1} - a_{2n}x_{2}^{k + 1} - \ldots - a_{\text{nn}}x_{n}^{k - 1} \right)$$

As in the first equation value of $x_{1}$ find by using previous value
of other variable in the equation and in 2<sup>nd</sup> equation value
of $x_{1}$is used which we find in first equation. Similarly, is the
case for last equation where all the new value of the variable is used.

<span class="underline">Example:</span>

$$\begin{matrix}
3x_{1} & x_{2} & x_{3} = 4 \\
2x_{1} & 5x_{2} & x_{3} = 6 \\
3x_{1} & 2x_{2} & 8x_{3} = 7 \\
\end{matrix}$$

The matrix $\begin{bmatrix}
3 & 1 & 1 \\
2 & 5 & 1 \\
3 & 2 & 8 \\
\end{bmatrix}$ is strictly dominant because magnitude of all the
diagonal entries are greater than the sum of all other entries of the
corresponding rows.

Then,

$$x_{1} = \frac{1}{3}\left( 4 - x_{2} - x_{3} \right)$$

$$x_{2} = \frac{1}{5}\left( 6 - 2x_{1} - x_{3} \right)$$

$$x_{3} = \frac{1}{8}\left( 7 - 3x_{1} - 2x_{2} \right)$$

Initially at $x_{1} = x_{2} = x_{3} = 0$

$$x_{1}^{1} = \frac{1}{3}\left( 4 - 0 - 0 \right) = 1.3333$$

$$x_{2}^{1} = \frac{1}{5}\left( 6 - \left( 2 \times 1.33 \right) - 0 \right) = 0.6667$$

$$x_{3}^{1} = \frac{1}{8}(7 - \left( 3 \times 1.3 \right) - \left( 2 \times 0.6667 \right)) = 0.2083$$

**Successive over relaxation method:**

<span class="underline">Introduction:</span>

Successive over relaxation is another method to solve the iterative
problems. In short it is called as SOR method. There is one parameter as
(w) is used which is equal to
$\frac{2}{\sqrt{1 - \rho\left( t_{j} \right)^{2}}}$

Where, $\rho\left( t_{j} \right)^{2}$ is the spectral radius which is
given as

$$\rho\left( t_{i} \right) = D^{- 1}(L + U)$$

There are some conditions for w:

-   If $w = 1$ this method is converted into Gauss-Seidel method.

-   If $w > 1\ $it is known as Over relaxation method.

-   If $w < 1$ it known as Under relaxation method.

-   The domain of w lies between the interval $(0,2)$. Only in this
    interval this method is converges otherwise diverges.

General formula for SOR method is given as

$$x_{i}^{k + 1} = x_{i}^{k} + \frac{w}{a_{\text{ij}}}\left\lbrack b_{i} - \sum_{j = 1}^{i - 1}{a_{\text{ij}}x_{j}^{k + 1}} - \sum_{j = 1}^{n}{a_{\text{ij}}x_{j}^{k}} \right\rbrack$$

***<span class="underline">Spline interpolation:</span>***

An interpolation in which special type of piecewise polynomial are
solve.

Spline interpolation is further divided into three methods which are as
follow;

-   Linear spline interpolation

-   Quadratic spline interpolation

-   Cubic spline interpolation

*<span class="underline">Linear spline interpolation:</span>*

The linear spline interpolation is used to represent set of line segment
between two data points.

Let
$\left( x_{0},y_{0} \right),\left( x_{1},y_{1} \right),\cdots,(x_{n},y_{n})$
be the data points. We should keep in value of x is in proper order
either it is in ascending order or in descending order. If the given
data is not in the particular order then first arrange in order such
that $x_{0} < x_{1} < x_{2} < \cdots < x_{n}$ or in the other way it is
represented as $x_{0} > x_{1} > x_{2} > \cdots > x_{n}$.

In linear spline interpolation linear equations are solve between
different intervals of the data point. Using this method equation are
form as

$$a_{1}x + b_{1}\text{\ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ }x_{0} < x < x_{1}$$

$$a_{2}x + b_{2}\text{\ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ }x_{1} < x < x_{2}$$

$$\ \  \vdots \  \vdots \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \  \vdots \  \vdots$$

$$a_{n}x + b_{n}\text{\ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ }x_{n - 1} < x < x_{n}$$

In this method the number of intervals is n-1 where n is the total
number of the data point and the number of equations formed from the
given intervals is same as total number of unknown variables found which
is equal to 2n.

Standard formula if linear spline interpolation is given as;

$$f_{i}\left( x \right) = y_{i} + \frac{y_{i + 1} - y_{i}}{x_{i + 1} - x_{i}}\left( x - x_{i} \right)\text{\ \ \ \ \ \ \ \ for\ }x_{i} \leq x \leq x_{i + 1}$$

Where, $i = 1,2,3,\cdots,n$

For $i = 1$ the above equation become;

$$f_{1}\left( x \right) = y_{1} + \frac{y_{2} - y_{1}}{x_{2 - x_{1}}}\left( x - x_{1} \right)\text{\ \ \ \ \ \ \ for\ }x_{1} \leq x \leq x_{2}$$

*<span class="underline">Quadratic spline interpolation:</span>*

In quadratic spline interpolation quadratic function are solve between
different intervals of the data point. Using this method equation are
form as

$$a_{1}x^{2} + b_{1}x + c_{1}\text{\ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ }x_{0} < x < x_{1}$$

$$a_{2}x^{2} + b_{2}x + c_{2}\text{\ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ }x_{1} < x < x_{2}$$

$$\ \  \vdots \  \vdots \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \  \vdots \  \vdots$$

$$a_{n}x^{2} + b_{n}x + c_{n}\text{\ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ }x_{n - 1} < x < x_{n}$$

In this method the number of intervals is n-1 where n is the total
number of the data point and the number of equations formed from the
given intervals is same as total number of unknown variables found which
is equal to 3n.

<span class="underline">Algorithm:</span>

-   Find all the equation on the given data point. Each equation is made
    by using two consecutive data points as

$$a_{1}x_{0}^{2} + b_{1}x_{0} + c_{1} = y_{0}$$

$$a_{2}x_{1}^{2} + b_{2}x_{1} + c_{2} = y_{1}$$

$$\ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \  \vdots \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \  \vdots \ \ \ \ \ \ \ \ \  \vdots \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ $$

$a_{n}x_{n}^{2} + b_{n}x_{n} + c_{n} = y_{n}$

Here the number of equations is 2n

-   As at the interior point the first derivative of the function is
    zero so

$$\frac{d}{\text{dx}}\left\lbrack a_{1}x_{1}^{2} + b_{1}x_{1} + c_{1} \right\rbrack = 0 = 2a_{1}x_{1} + b_{1}$$

$$\frac{d}{\text{dx}}\left\lbrack a_{2}x_{1}^{2} + b_{2}x_{1} + c_{2} \right\rbrack = 0 = 2a_{2}x_{1} + b_{2}$$

$$\ \  \vdots \vdots \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \  \vdots \vdots$$

$$\frac{d}{\text{dx}}\left\lbrack a_{n - 1}x_{n - 1}^{2} + b_{n - 1}x_{n - 1} + c_{n - 1} \right\rbrack = 0 = 2a_{n - 1}x_{n - 1} + b_{n - 1}$$

$$\frac{d}{\text{dx}}\left\lbrack a_{n - 1}x_{n}^{2} + b_{n - 1}x_{n} + c_{n} \right\rbrack = 0 = 2a_{n - 1}x_{n} + b_{n - 1}$$

> Which gives equations like

$$2a_{1}x_{1} + b_{1} = 2a_{2}x_{1} + b_{2}$$

$$2a_{1}x_{1} + b_{1} - 2a_{2}x_{1} - b_{2} = 0$$

$$\ \ \  \vdots \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \  \vdots$$

$$2a_{n - 1}x_{n - 1} + b_{n - 1} - 2a_{n - 1}x_{n} + b_{n - 1} = 0$$

> This gives n-1 equations.

-   If $|{x_{1} - x}_{0}| \leq |x_{n} - x_{n - 1}|$
    then$\text{\ \ \ \ }a_{1} = 0$.

This makes the first equation as linear. So, the number of equations is
3n.

*<span class="underline">Cubic spline interpolation:</span>*

It is most common piecewise polynomial approximation in which cubic
equations are formed between intervals. The equation in cubic spline has
both first and second derivative is the continuous function.

<span class="underline">Algorithm:</span>

-   Find all the equation on the given data point. Each equation is made
    by using two consecutive data points as

$$a_{1}x_{0}^{3} + b_{1}x_{0}^{2} + c_{1}x_{0} + d_{1} = y_{0}$$

$$a_{1}x_{1}^{3} + b_{1}x_{1}^{2} + c_{1}x_{1} + d_{2} = y_{1}$$

$$\ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \  \vdots \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \  \vdots \ \ \ \ \ \ \ \ \  \vdots \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ $$

$a_{n}x_{n}^{3} + b_{n}x_{n}^{2} + c_{n}x_{n} + d_{n} = y_{n}$

Here the number of equations is 2n

-   As at the interior point the first and second derivative of the
    function is zero so

<span class="underline">First derivative:</span>

$$\frac{d}{\text{dx}}\left\lbrack a_{1}x_{1}^{3} + b_{1}x_{1}^{2} + c_{1}x_{1} + d_{1} \right\rbrack = 3a_{1}x_{1}^{2} + 2b_{1}x_{1} + c_{1} = 0$$

$$\frac{d}{\text{dx}}\left\lbrack a_{2}x_{1}^{3} + b_{2}x_{1}^{2} + c_{2}x_{1} + d_{2} \right\rbrack = 3a_{2}x_{1}^{2} + 2b_{2}x_{1} + c_{2} = 0$$

$$\ \  \vdots \vdots \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \  \vdots \vdots$$

$$\frac{d}{\text{dx}}\left\lbrack a_{n - 1}x_{n - 1}^{3} + b_{n - 1}x_{n - 1}^{2} + c_{n - 1}x_{n - 1} + d_{n - 1} \right\rbrack = 3a_{n - 1}x_{n - 1}^{2} + 2b_{n - 1}x_{n - 1} + c_{n - 1} = 0$$

$$\frac{d}{\text{dx}}\left\lbrack a_{n - 1}x_{n}^{3} + b_{n - 1}x_{n}^{2} + c_{n}x_{n} + d_{n} \right\rbrack = 3a_{n - 1}x_{n}^{2} + b_{n - 1}x_{n} + c_{n - 1} = 0$$

> Which gives equations like

$$3a_{1}x_{1}^{2} + 2b_{1}x_{1} + c_{1} - 3a_{2}x_{1}^{2} - 2b_{2}x_{1} - c_{2} = 0$$

$$\ \  \vdots \vdots \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \  \vdots \vdots$$

$$3a_{n - 1}x_{n - 1}^{2} + 2b_{n - 1}x_{n - 1} + c_{n - 1} - 3a_{n - 1}x_{n}^{2} - b_{n - 1}x_{n} - c_{n - 1} = 0$$

<span class="underline">Second derivative:</span>

$$\frac{d}{\text{dx}}\left\lbrack a_{1}x_{1}^{3} + b_{1}x_{1}^{2} + c_{1}x_{1} + d_{1} \right\rbrack = \frac{d}{\text{dx}}\left\lbrack 3a_{1}x_{1}^{2} + 2b_{1}x_{1} + c_{1} \right\rbrack = 6a_{1}x_{1} + 2b_{1} = 0$$

$$\frac{d}{\text{dx}}\left\lbrack a_{2}x_{1}^{3} + b_{2}x_{1}^{2} + c_{2}x_{1} + d_{2} \right\rbrack = \frac{d}{\text{dx}}\left\lbrack 3a_{2}x_{1}^{2} + 2b_{2}x_{1} + c_{2} \right\rbrack = 6a_{2}x_{1} + 2b_{2} = 0$$

$$\ \  \vdots \vdots \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \  \vdots \vdots$$

$$\frac{d}{\text{dx}}\left\lbrack a_{n - 1}x_{n - 1}^{3} + b_{n - 1}x_{n - 1}^{2} + c_{n - 1}x_{n - 1} + d_{n - 1} \right\rbrack = \frac{d}{\text{dx}}\left\lbrack 3a_{n - 1}x_{n - 1}^{2} + 2b_{n - 1}x_{n - 1} + c_{n - 1} \right\rbrack = 6a_{n - 1}x_{n - 1} + 2b_{n - 1} = 0$$

$$\frac{d}{\text{dx}}\left\lbrack a_{n - 1}x_{n}^{3} + b_{n - 1}x_{n}^{2} + c_{n}x_{n} + d_{n} \right\rbrack = \frac{d}{\text{dx}}\left\lbrack 3a_{n - 1}x_{n}^{2} + b_{n - 1}x_{n} + c_{n - 1} \right\rbrack = 6a_{n - 1}x_{n} + 2b_{n - 1} = 0$$

> Which gives equations like

$$6a_{1}x_{1} + 2b_{1} = 6a_{2}x_{1} + 2b_{2}$$

$$6a_{1}x_{1} + 2b_{1} - 6a_{2}x_{1} - 2b_{2} = 0$$

$$6a_{n - 1}x_{n - 1} + 2b_{n - 1} - 6a_{n - 1}x_{n} - 2b_{n - 1} = 0$$

> This gives 2n-2 equations.

-   Now the total number of equations are 2n+2n-2=4n-2

This makes the first equation is quadratic and second equation is

***<span class="underline">Numerical Integration:</span>***

Integration is basically area under the curve bounded between the
interval (a, b). it is given by formula as

$$I = \int_{a}^{b}{f\left( x \right)\text{dx}}$$

Numerical integration is divided into further rule which are given
below:

-   Trapezoidal rule

-   Simpson’s $\frac{1}{3}^{\text{rd}}$rule

-   Simpson’s $\frac{3}{8}^{\text{th}}$rule

<span class="underline">Trapezoidal rule:</span>

Trapezoidal rule works in the linear state. The general formula is given
as:

$$I_{t} = \int_{x_{0}}^{x_{2}}{\text{hf}\left( p \right)\text{dp}}$$

$$I_{t} = h\int_{x_{0}}^{x_{2}}{f\left( p \right)\text{dp}}$$

$$I_{t} = \frac{h}{2}\left\lbrack f_{0} + 2\sum_{i = 2}^{n - 1}f_{i} + f_{n} \right\rbrack$$

Where, h= distance between two consecutive points

$$error = \frac{\left( b - a \right)^{2}}{12}f^{'}\left( \xi \right)\text{\ \ \ \ \ \ \ \ \ \ \ \ \ ξϵ}\left( a,\ b \right)$$

Generalized error:

$$E = \left| E.V - C.V \right| = \left| \int_{a}^{b}{f\left( x \right)\text{dx}} - error \right|$$

<span class="underline">Example:</span>

Consider the function $f\left( x \right) = x^{3} + 4x$ and find the
value of the function between the points $x_{0} = 0$ and $x_{2} = 1$
taking $h = 0.25$

$$I_{t} = \frac{h}{2}\left\lbrack f_{0} + 2\left( f_{1} + f_{2} \right) + f_{3} \right\rbrack$$

$$I_{t} = \frac{0.25}{2}\left\lbrack 0 + 2\left( 1.0156 + 3.4218 \right) + 5 \right\rbrack = 13.8748$$

**Numerical differentiation:**

Numerical differentiation methods are use to solve ordinary differential
equations. There are many methods in numerical differentiation some of
them are discussed below

-   Picard method

-   Taylor series method

-   Euler method

-   Modified Euler method

-   Improved Euler method

<span class="underline">Picard method:</span>

Picard method is also called Picard successive iteration method. In
Picard method the given ordinary differential equation is solve by using
integration of the function. The general formula for Picard method is
given as

$y_{n + 1} = y_{0} + \int_{x_{0}}^{x}{f(x,y_{n})}\text{dx}$

Where,

$n$ is the number of iterations.

<span class="underline">Algorithm:</span>

-   Start the iteration with taking $n = 0$ such that

$$y_{1} = y_{0} + \int_{x_{0}}^{x}{f(x,y_{0})}\text{dx}$$

Where,

Value of $y_{0}$ and $x_{0}$ is given in the question and the ordinary
differential equations is also given.

-   For 2<sup>nd</sup> iteration taking $n = 1$ then the formula becomes

$$y_{2} = y_{0} + \int_{x_{0}}^{x}{f(x,y_{1})}\text{dx}$$

-   The stopping criteria of this method is that when the integration
    becomes complicated then take the approximate solution of that
    iteration.

<span class="underline">Example:</span>

Consider the given ordinary differential equation
$\frac{\text{dy}}{\text{dx}} = x + y$ where $y\left( 0 \right) = 0$ and
find the value of $y(0.5)$ by using Picard method?

It is given that

$$\frac{\text{dy}}{\text{dx}} = x + y$$

$$x_{0} = 0$$

$$y_{0} = 0$$

$$y_{1} = y_{0} + \int_{0}^{x}{f\left( x,y_{0} \right)}\text{dx}$$

$$y_{1} = \frac{x^{2}}{2}$$

<span class="underline">Taylor series method:</span>

Taylor series is the infinite series in which derivative of
$n^{\text{th}}$power is calculated. The general formula of Taylor series
is given as

$$y\left( x \right) = y_{0} + \frac{\left( x - x_{0} \right){y^{'}}_{0}}{1!} + \frac{\left( x - x_{0} \right)^{2}{y^{''}}_{0}}{2!} + \frac{\left( x - x_{0} \right)^{3}{y^{'''}}_{0}}{3!} + \cdots\cdots$$

<span class="underline">Example:</span>

Consider the given ordinary differential equation
$\frac{\text{dy}}{\text{dx}} = x + y$ where $y\left( 0 \right) = 1$ and
find the value of $y(0.5)$ by using Taylor series method?

<span class="underline">Solution:</span>

$$\frac{\text{dy}}{\text{dx}} = x + y$$

$$y_{0} = 1$$

$$x_{0} = 0$$

$$y^{'} = x + y,{y'}_{0} = 1$$

$$y^{''} = y^{'},{y^{''}}_{0} = 1$$

$$y^{'''} = y^{''},{y^{'''}}_{0} = 1$$

$$y\left( x \right) = 1 + \frac{\left( x \right)\left( 1 \right)}{1!} + \frac{\left( x^{2} \right)\left( 1 \right)}{2!} + \frac{\left( x^{3} \right)\left( 1 \right)}{3!} + \cdots\cdots$$

$$y\left( x \right) = 1 + x + \frac{x^{2}}{2} + \frac{x^{3}}{6} + \cdots\cdots$$

Now,

For $x = 0.5$

$$y\left( 0.5 \right) = 1 + 0 + 0.125 + 0.0208 \approx 1.1458$$

<span class="underline">Euler’s method:</span>

The general formula of Euler method is given as

$$y_{n + 1} = y_{n} + hf\left( x_{n},y_{n} \right)$$

Where,

$n$ varies from 0,1,2,3,$\cdots$

$$h = \frac{final\ value - initial\ value}{\text{total\ steps\ performed}}$$

<span class="underline">Example:</span>

Consider the given ordinary differential equation
$\frac{\text{dy}}{\text{dx}} = x + y$ where $y\left( 0 \right) = 1$ and

<span class="underline">Solution:</span>

$$x_{0} = 0,\ \ \ \ x = 1.75$$

$$y_{0} = 1$$

$$h = \frac{x - x_{0}}{\text{steps}}$$

$$h = \frac{1.75 - 0}{3} = 0.5833$$

$$\frac{\text{dy}}{\text{dx}} = x + y$$

<span class="underline">Step 1:</span>

$for\ n = 0$

$$y_{1} = y_{0} + hf\left( x_{0},y_{0} \right)$$

$$y_{1} = 1 + \left( 0.5833 \right)\left( 0 + 1 \right)$$

$$y_{1} = 1.5833\ \ \ \ \ \ \ \ \ \ x_{1} = x_{0} + h = 0.5833$$

<span class="underline">Step 2:</span>

$for\ n = 1$

$$y_{2} = y_{1} + hf\left( x_{1},y_{1} \right)$$

$$y_{2} = 1.5833 + \left( 0.5833 \right)\left( 0.5833 + 1.5833 \right)$$

$$y_{2} = 2.8471\ \ \ \ \ \ \ \ \ \ x_{2} = x_{1} + h = 1.1666$$

<span class="underline">Modified Euler’s method:</span>
